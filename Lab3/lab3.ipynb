```python
import gaussian_beam_functions as gbf
```

```python
# Name of the file
filename = "./Data/Gaussian beam/10cm.bmp"

# Step 1: Load and visualize
image = gbf.extract_red(filename)
gbf.visualize_image(image, "Original Image", cmap="Reds")

# Step 2: Denoise
image_denoised = gbf.extract_red(filename, noise=True)
gbf.visualize_image(image_denoised, "Denoised Image", cmap="Reds")

# Step 3: Fit Gaussian
params = gbf.fit_gaussian(image_denoised)
gbf.visualize_fit(image_denoised, params)

# Step 5: Cross-Section Analysis
gbf.plot_cross_sections(image_denoised)

# Step 6: Beam Parameters
beam_params = gbf.calculate_beam_parameters(params)
print("Beam Parameters:", beam_params)

# Step 7: Fourier Analysis
fourier_magnitude = gbf.fourier_analysis(image_denoised)

# Step 8: Total Power
total_power = gbf.calculate_total_power(image_denoised)
print("Total Power:", total_power)
```
