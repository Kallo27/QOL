```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Constants
I_sat = 2.90e7  # Saturation intensity in W/m^2
g0 = 2.0  # Small signal gain coefficient
incident_intensity = np.linspace(0, 10 * I_sat, 500)  # Incident beam intensity range

# Saturated gain formula
gain = g0 / (1 + incident_intensity / I_sat)
```

```python
plt.figure(figsize=(8, 5))
plt.plot(incident_intensity, gain, label="Saturated gain", color="orange")
plt.axvline(I_sat, color="red", linestyle="--", label=r"$I_{\mathrm{sat}}$")
plt.title("Saturated gain as a function of incident intensity")
plt.xlabel("Incident intensity (W/m²)")
plt.ylabel(r"Saturated gain ($cm^{-1}$)")
plt.grid(alpha=0.3)
plt.legend()
plt.tight_layout()
plt.show()
```
